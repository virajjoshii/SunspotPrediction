```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import mean_absolute_error, r2_score

# Step 1: Preprocess the yearly dataset
def preprocess_yearly_data(df):
    """
    Cleans and prepares the yearly sunspot dataset for time series modeling.
    """
    # Convert year to datetime and filter valid years
    df['ds'] = pd.to_datetime(df[0], format='%Y', errors='coerce')
    df['y'] = pd.to_numeric(df[1], errors='coerce').replace(-1, np.nan)  # Replace invalid values with NaN
    df = df[['ds', 'y']].dropna()

    # Constrain the date range to valid datetime limits
    df = df[(df['ds'] >= pd.Timestamp('1900-01-01')) & (df['ds'] <= pd.Timestamp('2022-12-31'))]
    return df

# Step 2: Handle outliers
def handle_outliers(data):
    """
    Clips extreme outliers to avoid skewing model predictions.
    """
    q1 = data['y'].quantile(0.25)
    q3 = data['y'].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    data['y'] = data['y'].clip(lower_bound, upper_bound)
    return data

# Step 3: Build and forecast with Prophet
def build_fine_tuned_model(data, forecast_periods):
    """
    Builds a fine-tuned Prophet model and forecasts future values.
    """
    model = Prophet(
        growth="linear",
        changepoint_prior_scale=0.1,  # Increase sensitivity to trend changes
        seasonality_prior_scale=5.0  # Reduce overfitting to seasonality
    )
    model.add_seasonality(name='11_year_cycle', period=11, fourier_order=5)  # Add 11-year solar cycle seasonality
    model.fit(data)

    # Generate future dataframe and predict
    future = model.make_future_dataframe(periods=forecast_periods, freq='Y')
    forecast = model.predict(future)
    return model, forecast

# Step 4: Evaluate the model
def evaluate_model(y_true, y_pred):
    """
    Evaluates the model using MAE, SMAPE, and R² metrics.
    """
    epsilon = 1e-5  # Small constant to avoid division by zero
    y_true = np.where(y_true < epsilon, epsilon, y_true)  # Replace small actual values
    y_pred = np.where(y_pred < epsilon, epsilon, y_pred)  # Replace small predicted values

    # Calculate metrics
    mae = mean_absolute_error(y_true, y_pred)
    smape = np.mean(2 * np.abs(y_true - y_pred) / (np.abs(y_true) + np.abs(y_pred))) * 100
    r2 = r2_score(y_true, y_pred)

    return {"MAE": mae, "SMAPE": smape, "R²": r2}

# Step 5: Forecast for multiple horizons
def forecast_for_horizons(data, horizons):
    """
    Generates and displays forecasts for multiple horizons for yearly data.
    """
    results = {}
    for horizon in horizons:
        print(f"\nForecasting for {horizon} years...")

        # Build and forecast
        model, forecast = build_fine_tuned_model(data, horizon)

        # Plot the results
        model.plot(forecast)
        plt.title(f"Forecast for {horizon} Years")
        plt.xlabel("Year")
        plt.ylabel("Sunspot Count")
        plt.show()

        # Extract future predictions only
        future_predictions = forecast[['ds', 'yhat']].tail(horizon)
        print(f"Forecasted Values for {horizon} Years:")
        print(future_predictions)

        # Evaluate model
        evaluation = evaluate_model(data['y'], forecast['yhat'][:len(data)])
        results[horizon] = {"Evaluation Metrics": evaluation, "Forecast": future_predictions}
        print(f"Evaluation Metrics for {horizon} Years: {evaluation}")
    return results

# Load dataset
file_path = 'SN_y_tot_V2.0.csv'  # Replace with the actual file path
df = pd.read_csv(file_path, delimiter=';', header=None)

# Preprocess the yearly dataset
data = preprocess_yearly_data(df)
data = handle_outliers(data)

# Define forecast horizons
horizons = [1, 10, 20]  # Forecast for 1, 10, and 20 years

# Forecast and evaluate
forecast_results = forecast_for_horizons(data, horizons)
```
